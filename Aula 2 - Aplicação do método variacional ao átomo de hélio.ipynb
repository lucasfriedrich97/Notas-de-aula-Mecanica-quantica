## Aplicação do método variacional ao átomo de hélio

O hamiltoniano que descreve o átomo de hélio é dado por


$$
H = -\frac{\hbar^2}{2 m_{e}}( \nabla_{1}^2 + \nabla_{2}^2 ) - k e^2 \bigg( \frac{2}{r_1} + \frac{2}{r_2} - \frac{1}{|| \vec{r_1} - \vec{r_2} ||} \bigg)\tag{1}
$$


O valor aproximado da energia do átomo de hélio no estado fundamental é conhecido experimentalmente, e seu valor é

$$
E_{0}^{\text{exp}} \approx -78.935\,\text{eV}.\tag{2}
$$


A dificuldade em resolver este problema deve-se ao termo de repulsão elétron-elétron:

$$
U_{ee} =  \frac{ke^2}{|| \vec{r_1} - \vec{r_2} ||}.\tag{3}
$$

Sem esse termo, o hamiltoniano descrito pela Eq.(1) torna-se o hamiltoniano de dois átomos hidrogenoides, cuja solução é conhecida.

Poderíamos utilizar algumas abordagens para resolver este problema. São elas:


- Desconsiderar o termo $ U_{ee} $. Porém, nesse caso, a energia do estado fundamental $ E_{0} $ será aproximadamente:

$$
E_{(0)} \approx -108.85\,\text{eV}.\tag{4}
$$

- Calcular o valor médio $\langle  H \rangle $. Nesse caso, devemos propor uma função teste $\psi_{test}$ e, em seguida, calcular o valor médio. Uma função teste possível é dada por:


$$
\psi_{test} = \frac{8}{\pi a_{0}^{3}}e^{-2(r_1+r_2)}/a_{0},\tag{5}
$$
ou seja, $\psi_{test} = \psi_{test}(r_1,r_2) = \psi_{1,0,0}^{Z=2}(r_1)\psi_{1,0,0}^{Z=2}(r_2)$, onde 

$$
\psi_{1,0,0}^{Z=2}(r_j) = \sqrt{ \frac{8}{\pi a_{0}^3} } e^{-Zr_j/a_{0}},\tag{6}
$$
é a função de onda do estado fundamental de um átomo hidrogenóide. Nesse caso, ao calcularmos $ \langle H \rangle $, obteremos que:


$$
\langle  H \rangle  = -74.84\,\text{eV},\tag{7}
$$
o que está de acordo com o teorema variacional, visto que $\langle  H \rangle > E_{0}^{exp}$.


**OBS**: A obtenção detalhada desse valor pode ser encontrada em: https://github.com/jonasmaziero/mecanica_quantica/blob/main/notas_de_aula/40_2_Variational_Thm_He.ipynb

- Apesar de o valor obtido ao considerar $\langle  H \rangle$ ser razoavelmente bom, ele ainda pode ser melhorado. Para isso, consideramos a adição de um **parâmetro variacional** ao calcularmos a média $\langle  H \rangle$.



## Aplicando o método variacional

Para usarmos o método variacional, lembremos que devemos seguir um passo a passo, tal que:

- Iniciamos propondo uma função teste.

- Calculamos o valor médio $\langle  H \rangle$.

- Otimizamos o parâmetro variacional.


### Função de onda teste

Uma função de onda teste possível é:

$$\psi_{Z} = \psi_{Z}(r_1,r_2) = \psi_{1,0,0}^{Z}(r_1)\psi_{1,0,0}^{Z}(r_2),\tag{8}$$
com $\psi_{1,0,0}^{Z}(r_j)$ definido como

$$
\psi_{1,0,0}^{Z}(r_j) = \sqrt{ \frac{Z^3}{\pi a_{0}^3} } e^{-Zr_j/a_{0}},\tag{9}
$$

Em outras palavras, iremos considerar uma função teste obtida a partir do produto da função de onda do átomo hidrogenóide, onde agora, em vez de usarmos $ Z = 2 $, $ Z $ será nosso parâmetro variacional. Neste caso, nossa função teste será escrita como:

$$
\psi_{Z} = \frac{Z^3}{\pi a_{0}^3}e^{-Z(r_1+r_2)/a_0}\tag{10}
$$

**Interpretação desta nova função de onda**: Na figura abaixo, temos a representação do átomo de hélio, onde em:

A) Temos o núcleo na origem do nosso referencial e os dois elétrons orbitando em torno deste núcleo. Ao usarmos $ \psi_{\text{test}} $ para calcular o valor médio de $ H $, usamos $ Z = 2 $, pois este é o valor de $ Z $ para o átomo de hélio. Porém, isso implica que cada elétron está 'vendo' a carga do núcleo sem interferência do outro elétron, ou seja, é como se, para cada elétron, o outro não existisse ou não influenciasse a sua interação com o núcleo.

B) Nova representação do átomo de hélio. Agora, cada elétron não 'enxerga' o núcleo como algo isolado do outro elétron. Na verdade, agora, cada elétron vai 'enxergar' o núcleo e o outro elétron como um único ente.

C) Agora, para cada elétron, teremos um núcleo com uma carga efetiva onde agora o elétron 'vê' o outro elétron e o núcleo como uma coisa só.


### Valor médio

Definida a função teste, agora devemos calcular o valor médio $ \langle H \rangle_{\psi_{Z}} $. Aqui, podemos nos fazer a seguinte pergunta: podemos calcular o valor médio usando diretamente a função $\psi_{Z}$? A resposta é sim, nada nos impediria de fazer isso.

Todavia, para tornar as contas que deveremos fazer mais simples, podemos fazer as seguintes observações:

- A Eq.(9) é a função de onda para o estado fundamental do átomo hidrogenóide.


- O hamiltoniano do átomo hidrogenóide é definido como:

$$
H = - \frac{\hbar^2 \nabla^2}{2m} - \frac{Zke^2}{r}.\tag{11}
$$

- As energias em função de $Z$ são dadas por:

$$
E_{n}^{Z} = - \frac{Z^2|E_1|}{n^2}, \quad \text{com} \quad |E_1| = 13.6eV.\tag{12}
$$
Logo, para $n = 1$, temos que:

$$
E_{1}^{Z} = E^{Z} = - Z^2|E_1|.\tag{13}
$$

**Por que isso é relevante?** Quando aplicamos o **método variacional** para analisar um sistema, o seu **hamiltoniano não é alterado**, somente a função de onda é que é mudada. Todavia, isso não nos impede de reescrevermos o hamiltoniano de uma forma que torne a solução do problema mais fácil.

Portanto, para facilitar o cálculo de $ \langle H \rangle_{\psi_Z} $, iremos reescrever o hamiltoniano, descrito pela Eq.(1), de modo a facilitar nossas contas. Para isso, iremos tentar reescrever este hamiltoniano de modo que apareça algo similar ao hamiltoniano descrito na Eq.(11). Assim sendo, podemos reescrever o hamiltoniano dado pela Eq.(1), da seguinte forma:


$$
H = \bigg\{ -\frac{\hbar^2}{2 m_{e}}( \nabla_{1}^2 + \nabla_{2}^2 ) - k e^2 \bigg( \frac{2}{r_1} + \frac{2}{r_2} - \frac{1}{|| \vec{r_1} - \vec{r_2} ||} \bigg) \bigg\} + \frac{Ze^2k}{r_1} - \frac{Ze^2k}{r_1} + \frac{Ze^2k}{r_2} - \frac{Ze^2k}{r_2},\tag{14}
$$
ou seja, somamos e subtraímos $\frac{Ze^2k}{r_1}$ e $\frac{Ze^2k}{r_2}$, o que não afeta em nada o hamiltoniano do sistema. Todavia, ao fazermos isso, podemos reordenar os termos de modo que obtemos:


\begin{equation}
\begin{aligned}
H & = \bigg [ -\frac{\hbar^2 \nabla_{1}^2}{2m} - \frac{Ze^2k}{r_1} \bigg] + \bigg [ -\frac{\hbar^2 \nabla_{2}^2}{2m} - \frac{Ze^2k}{r_2} \bigg] +  (Z-2)ke^2 \bigg[ \frac{1}{r_1} + \frac{1}{r_2} \bigg] + \frac{ke^2}{||\vec{r_1} - \vec{r_2}||} \\
& = H_{Z}^{1} + H_{Z}^{2} + (Z-2)ke^2 \bigg[ \frac{1}{r_1} + \frac{1}{r_2} \bigg] + \frac{ke^2}{||\vec{r_1} - \vec{r_2}||}
\end{aligned}
\tag{15}
\end{equation}


Agora, que temos $H$, podemos calcular seu valor médio, tal que:

$$
\langle H \rangle_{\psi_Z} = \langle H_{Z}^{1} \rangle_{\psi_Z} + \langle H_{Z}^2 \rangle_{\psi_Z} + 
(Z-2)ke^2 \bigg[ \bigg\langle \frac{1}{r_1} \bigg\rangle_{\psi_Z} + \bigg\langle \frac{1}{r_2} \bigg\rangle_{\psi_Z} \bigg] + \langle U_{ee} \rangle_{\psi_Z},\tag{16}
$$
onde 

$$
U_{ee} = \frac{ke^2}{||\vec{r_1} - \vec{r_2}||}\tag{17}
$$

-------------

- Vamos iniciar calculando $ \langle H_{Z}^{1} \rangle_{\psi_Z}$.

Usando a função teste, definida na Eq.(8), temos que:

\begin{equation}
\begin{aligned}
 \langle H_{Z}^{1} \rangle_{\psi_Z} & = \int \int \psi_{Z}^{*} H_{Z}^{1} \psi_{Z} d^3r_1d^3r_2 \\
 & = \int \int \psi_{1,0,0}^{Z}(r_1)^{*}\psi_{1,0,0}^{Z}(r_2)^{*} H_{Z}^{1} \psi_{1,0,0}^{Z}(r_1)\psi_{1,0,0}^{Z}(r_2) d^3r_1d^3r_2.
\end{aligned}
\tag{18}
\end{equation}

Neste ponto, observe o seguinte: $H_{Z}^{1}$ é o hamiltoniano associado ao elétron descrito por $\vec{r_1}$. Logo, ele atua somente na função de onda associada a este elétron, ou seja, na função $\psi_{1,0,0}^{Z}(r_1)$. Portanto, podemos reescrever a Eq.(18), como segue:

\begin{equation}
\begin{aligned}
\langle H_{Z}^{1} \rangle_{\psi_Z} & = \int \psi_{1,0,0}^{Z}(r_1)^{*}H_{Z}^{1} \psi_{1,0,0}^{Z}(r_1) d^3r_1 \int \psi_{1,0,0}^{Z}(r_2)^{*} \psi_{1,0,0}^{Z}(r_2) d^3r_2 \\
& = \int \psi_{1,0,0}^{Z}(r_1)^{*}H_{Z}^{1} \psi_{1,0,0}^{Z}(r_1) d^3r_1 = \langle H_{Z}^{1} \rangle_{\psi_{1,0,0}^{Z}(r_1)}.
\end{aligned}
\tag{19}
\end{equation}

Agora, ao analisarmos o termo $\langle H_{Z}^{1} \rangle_{\psi_{1,0,0}^{Z}(r_1)}$, veremos que isso nada mais é do que o valor médio do hamiltoniano do átomo hidrogenóide usando a própria função de onda do estado fundamental do átomo hidrogenóide. Logo


$$
\langle H_{Z}^{1} \rangle_{\psi_Z} = \langle H_{Z}^{1} \rangle_{\psi_{1,0,0}^{Z}(r_1)} = E^{Z} = - Z^2|E_1|.\tag{20}
$$

**OBS**: A mesma análise pode ser feita para $\langle H_{Z}^{2} \rangle_{\psi_{Z}}$, de modo que também obteremos $\langle H_{Z}^{2} \rangle_{\psi_Z} = \langle H_{Z}^{2} \rangle_{\psi_{1,0,0}^{Z}(r_2)} = E^{Z} = - Z^2|E_1|$.


Assim, da Eq.(16), obtemos que:

$$
\langle H \rangle_{\psi_Z} = -2Z^2|E_1| + 
(Z-2)ke^2 \bigg[ \bigg\langle \frac{1}{r_1} \bigg\rangle_{\psi_Z} + \bigg\langle \frac{1}{r_2} \bigg\rangle_{\psi_Z} \bigg] + \langle U_{ee} \rangle_{\psi_Z}.\tag{21}
$$

Agora, se usarmos que
$$  \bigg\langle \frac{1}{r_1} \bigg\rangle_{\psi_Z} =  \bigg\langle \frac{1}{r_2} \bigg\rangle_{\psi_Z} =  \bigg\langle \frac{1}{r} \bigg\rangle_{\psi_Z} = \frac{Z}{a_0}, \tag{22} $$
iremos obter

$$
\langle H \rangle_{\psi_Z} = -2Z^2|E_1| + 
2(Z-2)ke^2 \frac{Z}{a_0} + \langle U_{ee} \rangle_{\psi_Z}.\tag{23}
$$

**Atividade**: Demonstre as igualdades da Eq.(22).

Agora, para finalizarmos, devemos obter uma expressão para $\langle U_{ee} \rangle_{\psi_Z}$. Assim sendo, iniciamos escrevendo

\begin{equation}
\begin{aligned}
\langle U_{ee} \rangle_{\psi_{Z}} & = \int \int d^3r_1d^3r_2 \psi_{Z}^{*} \frac{ke^2}{||\vec{r_1}-\vec{r_2}||}\psi_{Z} \\
& = ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2 \int \int d^3r_1d^3r_2 \frac{ e^{-2Z(r_1+r_2)/a_0} }{||\vec{r_1}-\vec{r_2}||},
\end{aligned}
\tag{24}
\end{equation}

onde, se usarmos que
\begin{equation}
\begin{aligned}
|| \vec{r_1} - \vec{r_2} || & = \sqrt{ (\vec{r_1} - \vec{r_2}) \cdot (\vec{r_1} - \vec{r_2})  } \\
& = \sqrt{ r_1^2 + r_2^2 - 2 \vec{r_1} \cdot \vec{r_2} } \\
& = \sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 },
\end{aligned}
\tag{25}
\end{equation}
com $\theta_2$ sendo o ângulo entre os vetores $\vec{r_1}$ e $\vec{r_2}$. Então, da Eq.(24), obteremos

\begin{equation}
\begin{aligned}
\langle U_{ee} \rangle_{\psi_{Z}} 
& = ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2 \int \int d^3r_1d^3r_2 \frac{ e^{-2Z(r_1+r_2)/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }} \\
& = ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2 \int d^3r_1 e^{-2Zr_1/a_0} \int d^3r_2 \frac{ e^{-2Zr_2/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }}.
\end{aligned}
\tag{26}
\end{equation}

Agora, vamos resolver primeiro a integral em relação a $r_2$. Para tal, iremos mudar esta integral para coordenadas esféricas, de modo que

\begin{equation}
\begin{aligned}
\int d^3r_2 \frac{ e^{-2Zr_2/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }} & = \int_{0}^{\infty} \int_{0}^{\pi} \int_{0}^{2\pi} r_2^2 \sin \theta_2 dr_2 d \theta_2 d \phi_2 \frac{ e^{-2Zr_2/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }} \\
& =  2\pi \int_{0}^{\infty} r_2^2 dr_2 e^{-2Zr_2/a_0} \int_{0}^{\pi} \frac{ d \theta_2 sin \theta_2 }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }}.
\end{aligned}
\tag{27}
\end{equation}


Agora, temos que resolver uma integral em relação a $r_2$ e outra em relação a $\theta_2$. Vamos começar por esta última, usando para isso a seguinte mudança de variáveis: $ u = r_1^2 + r_2^2 - 2r_1r_2 \cos \theta_2 $, de modo que $d u = 2 r_1 r_2 \sin \theta_2 \, d\theta_2$, logo


\begin{equation}
\begin{aligned}
\int_{0}^{\pi} \frac{ d \theta_2 sin \theta_2 }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }} & = \frac{1}{2r_1r_2} \int u^{-1/2}du \\ &
\vdots  \\&
= \frac{ \sqrt{ r_1^2+r_2^2-2r_1r_2 \cos \theta_2 } }{r_1r_2}\bigg|_{0}^{\pi} \\
& = \frac{1}{r_1r_2} \big[ \sqrt{r_1^2+r_2^2+2r_1r_2} - \sqrt{r_1^2+r_2^2-2r_1r_2} \big] \\
& = \frac{1}{r_1r_2} \big[ \sqrt{(r_1+r_2)^2} - \sqrt{(r_1-r_2)^2} \big] \\
& = \frac{ r_1+r_2-|r_1-r_2| }{r_1r_2}.
\end{aligned}
\tag{28}
\end{equation}

Assim, da Eq.(27), obtemos que

$$
\int d^3r_2 \frac{ e^{-2Zr_2/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }}  =  2\pi \int_{0}^{\infty} r_2^2 dr_2 e^{-2Zr_2/a_0}\frac{ r_1+r_2-|r_1-r_2| }{r_1r_2},\tag{29}
$$
ou ainda, como

$$
\frac{ r_1+r_2-|r_1-r_2| }{r_1r_2} = \frac{2}{r_1}, \quad \text{se} \quad r_1 \geqslant  r_2\tag{30}
$$
e
$$
\frac{ r_1+r_2-|r_1-r_2| }{r_1r_2} = \frac{2}{r_2}, \quad \text{se} \quad r_1 <  r_2.\tag{31}
$$

então

\begin{equation}
\begin{aligned}
\int d^3r_2 \frac{ e^{-2Zr_2/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }} & =  2\pi \bigg[ \int_{0}^{r_1}dr_2 r_2^2 e^{-2Zr_2/a_0} \frac{2}{r_1}  + \int_{r_1}^{\infty} dr_2 r_2^2 e^{-2Zr_2/a_0} \frac{2}{r_2} \bigg] \\
& = \frac{4\pi}{r_1} \bigg[ \int_{0}^{r_1}dr_2 r_2^2 e^{-2Zr_2/a_0}  + r_1 \int_{r_1}^{\infty} dr_2 r_2 e^{-2Zr_2/a_0}  \bigg].
\end{aligned}
\tag{32}
\end{equation}

Agora, para resolvermos estas duas integrais, iremos usar as seguintes igualdades

$$
\int x e^{-\alpha x}dx = -\frac{1}{\alpha}\bigg( 1+\frac{1}{\alpha} \bigg)e^{-\alpha x}\tag{33}
$$
e
$$
\int x^2e^{-\alpha x}dx = -\frac{1}{\alpha} \bigg[ x^2 + \frac{2}{\alpha}\bigg( x + \frac{1}{\alpha} \bigg)  \bigg]e^{-\alpha x}.\tag{34}
$$

Portanto, para a primeira integral, temos que:


\begin{equation}
\begin{aligned}
\int_{0}^{r_1}dr_2 r_2^2 e^{-2Zr_2/a_0} & = -\frac{a_0}{2Z} \bigg[ r_2^2 + \frac{2a_0}{2Z} \bigg( r_2 + \frac{a_0}{2Z} \bigg) \bigg] e^{-2Zr_2/a_0} \bigg|_{0}^{r_1} \\
& =  -\frac{a_0}{2Z} \bigg[ r_1^2 + \frac{a_0}{Z} \bigg( r_1 + \frac{a_0}{2Z} \bigg) \bigg] e^{-2Zr_1/a_0} 
+  \frac{a_0}{2Z} \bigg[ 0 + \frac{a_0}{Z} \bigg( 0 + \frac{a_0}{2Z} \bigg) \bigg] e^{0} \\
& = -\frac{a_0}{2Z} \bigg[ r_1^2 + \frac{a_0}{Z} \bigg( r_1 + \frac{a_0}{2Z} \bigg) \bigg] e^{-2Zr_1/a_0} + \frac{1}{4}\bigg( \frac{a_0}{Z} \bigg)^3.
\end{aligned}
\tag{35}
\end{equation}

Já para a segunda, temos que:


\begin{equation}
\begin{aligned}
\int_{r_1}^{\infty} dr_2 r_2 e^{-2Zr_2/a_0} & = - \frac{a_0}{2Z}\bigg( r_2 + \frac{a_0}{2Z} \bigg)e^{-2Zr_2/a_0} \bigg|_{r_1}^{\infty} \\
& = 0 + \frac{a_0}{2Z}\bigg( r_1 + \frac{a_0}{2Z} \bigg)e^{-2Zr_1/a_0} \\
& = \frac{a_0}{2Z}\bigg( r_1 + \frac{a_0}{2Z} \bigg)e^{-2Zr_1/a_0}.
\end{aligned}
\tag{36}
\end{equation}

Assim, substituindo esses dois resultados na Eq.(32), e fazendo as devidas manipulações, obteremos

$$
\int d^3r_2 \frac{ e^{-2Zr_2/a_0} }{\sqrt{ r_1^2 + r_2^2 - 2 r_1 r_2 \cos\theta_2 }}  = \frac{\pi}{r_1}\bigg( \frac{a_0}{Z} \bigg)^3 \bigg[ 1 - \bigg( \frac{Zr_1}{a_0} + 1 \bigg)e^{-2Zr_1/a_0} \bigg].\tag{37}
$$

Agora, voltemos para a Eq.(26), de modo que, usando o resultado obtido na Eq.(37), iremos obter

\begin{align}
\langle U_{ee} \rangle_{\psi_{Z}} 
& = \pi ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2  \bigg( \frac{a_0}{Z} \bigg)^3  \int d^3r_1  \frac{ e^{-2Zr_1/a_0} }{ r_1 }\bigg[ 1 - \bigg( \frac{Zr_1}{a_0} + 1 \bigg)e^{-2Zr_1/a_0} \bigg],\tag{38}
\end{align}

ou, em coordenadas esféricas

\begin{align}
\langle U_{ee} \rangle_{\psi_{Z}} 
& = \pi ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2  \bigg( \frac{a_0}{Z} \bigg)^3  \int_{0}^{2\pi} d \phi_1 \int_{0}^{\pi} \sin \theta_1 d \theta_1 \int_{0}^{\infty} r_1^2  \frac{ e^{-2Zr_1/a_0} }{ r_1 }\bigg[ 1 - \bigg( \frac{Zr_1}{a_0} + 1 \bigg)e^{-2Zr_1/a_0} \bigg]dr_1.\tag{39}
\end{align}



As duas primeiras integrais são de fácil solução. Assim

\begin{align}
\langle U_{ee} \rangle_{\psi_{Z}} 
& = 4 \pi^2 ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2  \bigg( \frac{a_0}{Z} \bigg)^3  \int_{0}^{\infty} r_1   e^{-2Zr_1/a_0} \bigg[ 1 - \bigg( \frac{Zr_1}{a_0} + 1 \bigg)e^{-2Zr_1/a_0} \bigg]dr_1,\tag{40}
\end{align}

ou ainda 


\begin{align}
\langle U_{ee} \rangle_{\psi_{Z}} 
& = 4 \pi^2 ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2  \bigg( \frac{a_0}{Z} \bigg)^3 \bigg[ \int_{0}^{\infty} r_1e^{-2Zr_1/a_0} dr_1  - \frac{Z}{a_0} \int_{0}^{\infty} r_1^2e^{-4Zr_1/a_0}dr_1 -\int_{0}^{\infty} r_1 e^{-4Zr_1/a_0}dr_1  \bigg].\tag{41}
\end{align}

Para resolvermos estas três integrais, recorremos novamente às identidades apresentadas na Eq.(33) e na Eq.(34), de modo que:

$$
 \int_{0}^{\infty} r_1e^{-2Zr_1/a_0} = \cdots = \bigg(\frac{a_0}{2Z}\bigg)^2,\tag{42}
$$

$$
\int_{0}^{\infty} r_1^2e^{-4Zr_1/a_0}dr_1 = \cdots = \frac{1}{32}\bigg(\frac{a_0}{Z}\bigg)^2\tag{43}
$$

e

$$
\int_{0}^{\infty} r_1 e^{-4Zr_1/a_0}dr_1 = \cdots = \bigg(\frac{a_0}{4Z}\bigg)^2.\tag{44}
$$

Portanto, da Eq.(41), obtemos que:

\begin{equation}
\begin{aligned}
\langle U_{ee} \rangle_{\psi_{Z}} 
& = 4 \pi^2 ke^2 \bigg( \frac{Z^3}{\pi a_{0}^3} \bigg)^2  \bigg( \frac{a_0}{Z} \bigg)^3 \bigg[ \bigg(\frac{a_0}{2Z}\bigg)^2 - \frac{Z}{a_0} \frac{1}{32}\bigg(\frac{a_0}{Z}\bigg)^2 - \bigg(\frac{a_0}{4Z}\bigg)^2 \bigg] \\
& \vdots \\
& = \frac{Z}{2} \frac{5ke^2}{4a_0}.
\end{aligned}
\tag{45}
\end{equation}

Agora que já temos uma expressão para $\langle U_{ee} \rangle_{\psi_{Z}}$, podemos voltar para a Eq.(23). Assim, usando a expressão para $\langle U_{ee} \rangle_{\psi_{Z}}$, obtida na Eq.(45), obteremos que

$$
\langle H \rangle_{\psi_Z} = -2Z^2|E_1| + 
2(Z-2)ke^2 \frac{Z}{a_0} + \frac{Z}{2} \frac{5ke^2}{4a_0}.\tag{46}
$$

Se agora usarmos $E_{1}^{(0)} = -|E_1| = - \frac{ke^2}{2a_0}$, então, fazendo as devidas manipulações, teremos que
$$
\langle H \rangle_{\psi_Z} = \bigg( -2Z^2 + \frac{27}{4}Z \bigg)E_{1}^{(0)}\tag{47}
$$

### Otimização do parâmetro variacional

Agora que temos uma expressão para $\langle H \rangle_{\psi_Z}$, devemos achar $Z_{opt}$, que minimize $\langle H \rangle_{\psi_Z}$. Para tal, procedemos da seguinte forma:

$$
\frac{\partial \langle H \rangle_{\psi_Z}}{\partial Z} = -4Z + \frac{27}{4} = 0 \longrightarrow Z_{opt} = \frac{27}{16} = 1.67875.\tag{48}
$$

Portanto, usando este valor na Eq.(47), obtemos que

$$
\langle H \rangle_{\psi_{Z_{opt}}} \approx -77.4 \, \text{eV}.\tag{49}
$$

Que é um valor mais próximo ao experimental, cujo valor é $E_{0}^{\text{exp}} \approx - 78.935 \, \text{eV}$.
